```python

import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
df=pd.read_csv('milk_production.csv')
df=df.applymap(lambda x:x.strip() if isinstance(x,str) else x)
df.fillna(0,inplace=True)
df.head()

```

```python
df['total_milk']=df['Cow Milk-2013-14']+df['Boffalo Milk-2013-14']+df['Goat Milk-2013-14']
df.head()
```

```python
print('  The state has produced max milk in year 2013-14')
df[df['total_milk']==df['total_milk'].max()]['State/ UT Name']
```

```python
df['total milk 2010-15']=df['Cow Milk-2010-11']+df['Cow Milk-2011-12']+df['Cow Milk-2013-14']+df['Cow Milk-2014-15']+df['Boffalo Milk-2010-11']+df['Boffalo Milk-2011-12']+df['Boffalo Milk-2013-14']+df['Boffalo Milk-2014-15']+df['Goat Milk-2011-12']+df['Goat Milk-2013-14']+df['Goat Milk-2014-15']
df.head()
```

```python
print('top 5 milk producing states in each year from 2010-15')
q=df.sort_values('total milk 2010-15',ascending=False)['State/ UT Name']
q.head()
```

```python
df.mean(axis=1)
```

```python
print('avg cow milk production for state 1')
p=df.iloc[0:1,2:7].mean(axis=1)

q=DataFrame(p)
q
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df['State/ UT Name']
sizes = df['total milk 2010-15']
#explode = (0, 0, 0, 0)
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
plt.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=25)
ax1.axis('equal')
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

plt.show()
```
