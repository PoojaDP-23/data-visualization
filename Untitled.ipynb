```python


import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
df1=pd.read_csv('oil_reserves.csv')
df1=df1.applymap(lambda x:x.strip() if isinstance(x,str) else x)
df1.fillna(0,inplace=True)
df1
```

```python
#1. Which region has max oil reserves from 2014 to 2016?
print('The region with maximum oil')
df1['total oil']=df1['2014']+df1['2015']+df1['2016']
df1[df1['total oil']==df1['total oil'].max()][['Region','total oil']]
```

```python
#3. List the countries having maximum oil reserves in each region.
df1.sort_values('total oil',ascending=False)[['Country/ Region','total oil']]
```

```python
#2. With suitable visualization show the oil reserves of India from 2010 to 2016.
p=df1.iloc[48:49,1:7]
p
```

```python
#5. Find the world’s top 5 countries having maximum oil reserves in the 2015. Show their % of share in total oil reserve for the same
df1.sort_values('2015',ascending=False)[['Country/ Region','2015']].head()
```

```python
#4. Draw suitable visualization which shows the share of oil reserves by each region.
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df1['Country/ Region']
sizes = df1['total oil']
#explode = (0, 0, 0, 0)
#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
plt.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=125)
ax1.axis('equal')
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

plt.show()
```
