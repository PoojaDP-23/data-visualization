```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel('MONTHLY_ACTUAL_RAINFALL_FROM_2009_TO_2011.XLS', index_col = False)


#plot data
plt.plot(data.loc['Coastal Karnataka'], data['B'], linestyle="dotted", marker='^', color="red", label = 'Product 1')

plt.plot(data.loc['North Interior Karnataka'], data['B'], linestyle="solid", marker='o', color="green", label = 'Product 2')

plt.plot(data.loc['South Interior Karnataka'], data['B'], linestyle="dashed", marker='<', color="blue", label = 'Product 3')



#labels
plt.xlabel("divison")
plt.ylabel("year 2002")

# add legend
plt.legend()

#title
plt.title("rainfall")

#show plot
plt.show()                             
```
